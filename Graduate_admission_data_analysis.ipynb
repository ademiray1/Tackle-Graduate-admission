####1. Load Data

Import numpy, pandas and matplotlib libraries, and load data into Pandas dataframe.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#those are the three main libraries we are going to use

#Let's read the data

df = pd.read_csv('Admission_predict_data.csv')

print(df.shape)

df.info()

#Giving 500 rows and 9 columns and all columns are numerical
```

```python
#### 2. Clean up Columns

Delete the first column (serial num) since Pandas already assigns an id to each row.

Rename the columns to simpler names. Print first 5 rows.
```

```python
df = df.drop(['Serial No.'], axis=1) # axis =1 means we are going to delete a column

#creating dictionary for renaming the column names
new_names = {'GRE Score':'gre', 'TOEFL Score':'toefl','University Rating':'rating', 'SOP':'sop', 'LOR':'lor', 'CGPA':'gpa', 'Research':'research', 'Chance of Admit ':'chance'}

#sop: statement of purpose
#lor: letters of recommendation
#research: is binary 1 yes 0 no
#chance : each of has correleation

df.rename(columns=new_names,inplace=True)
df.head() # first five lines of the data
```

#### 3. Scope out Data

Show boxplots to see high-level distribution of main columns
Use pandas.describe() to see high-level distribution of data

```python
plt.figure(1, figsize=(14,4)) #first one is height second is width
for i in range(1,6):
    plt.subplot(1,5,i) # here 1 means row, 5 columns and i ist i'th item
    plt.boxplot(df[df.columns[i]]) #we set the data using whichever column we want we grab
    plt.title(df.columns[i])
plt.show()
df.describe() # statistical summary of our data frame.
```

#### 4. Detailed Data Distribution

Plot histograms for main columns to show detailed distribution of data

```python
df[['gre','toefl','rating','gpa','research','chance']].hist(figsize=(14,9),bins=20,linewidth='1',edgecolor='k',grid=True)
#bins basicly means how many columns we want in this chart!

plt.show()

#We can see that most of the data normall distributed

#research is not normal distributed because its a binary data
```

#### 5. Show correlation with Chance of Acceptance

Calculate correlation between each data column and Chance of Acceptance

Here we can see GPA, GRE score and TOEFL score are the most important feaures because they have the best correlation with acceptance.

Research appears to be the least important feature.




```python
correlation = df.corr()['chance']
plt.bar(df.columns, correlation, color = 'gray') # first one is columns second is y
plt.show()
print(correlation)

#GPA is the tallest one. That's probably the most important factor in acceptance followed by GRE and TOELF

```

#### 6. Plot Relations between GRE & TOEFL, and GPA & GRE

We can see a strong correlation between TOEFL and GRE scores -- people scoring high on one probably scored high on the other.
And we see a strong correlation between GPA and GRE -- people with a high GPA probably scored high on the GRE.



```python
df.plot(kind='scatter', x='gre', y='toefl', color='green')
plt.xlabel("GRE")
plt.ylabel("TOEFL")
plt.show()

df.plot(kind='scatter', x='gpa', y='gre', color='red')
plt.xlabel("GPA")
plt.ylabel("GRE")
plt.show()

#Basicly tighter the cluster is around the centerline the tighter the correlation is between them

#there is a clear relationship between GRE and GPA 
```

```python
#### 7. Plot Relations between chance of acceptence and other Features



for i in range(0,7):
    df.plot(kind='scatter', x='chance', y=df.columns[i]) 
    plt.xlabel("Chance")
    plt.ylabel(df.columns[i])
    plt.show()
    
    #rating of the school vs Change does not have a good relationship
```
